```python
import sklearn.datasets
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import warnings
import numpy as np
from matplotlib import pyplot as plt
warnings.filterwarnings('ignore')
```

```python
data = sklearn.datasets.load_digits()
```

```python
X = data.data
y = data.target
```

```python
clf = DecisionTreeClassifier()
cross_val_score(clf, X=X, y=y, cv=10).mean()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bcf = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_features=8), n_estimators=100)
```

```python
cross_val_score(bcf, X=X, y=y, cv=10).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randForest = RandomForestClassifier(n_estimators=6, max_features=8)
cross_val_score(randForest, X=X, y=y, cv=10).mean()
```

```python
plt.figure(figsize=(8, 5))
a = np.arange(5, 101, 5)
f = []
print(a)
for i in range(len(a)):
    r = cross_val_score(randForest, X=X, y=y, cv=a[i]).mean()
    f.append(r)
plt.plot(a, f)
```

```python
plt.figure(figsize=(8, 5))
a = np.arange(100, 130, 5)
f = []
print(a)
for i in range(len(a)):
    r = cross_val_score(randForest, X=X, y=y, cv=a[i]).mean()
    f.append(r)
plt.plot(a, f)
```
