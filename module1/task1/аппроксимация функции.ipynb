```python
import scipy
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 15, 0.1) # Создадим массив от 1 до 15 с шагом 0.1
y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
plt.plot(x, y)
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)


y1 = f(1)
y15 = f(15)



y4 = f(4)
y10 = f(10)
y8 = f(8)
print(y1, y15)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([y1, y15])
x = scipy.linalg.solve(a, b)
print(a)
print("y = ", b)
print("коэффициенты - ", x)
```

```python
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([y1, y8, y15])
x = scipy.linalg.solve(a, b)
print(a)
print("y = ", b)
print(x)
a1 = np.array([1, 8, 15])
b1 = np.array([y1, y8, y15])
plt.plot(a1, b1)
```

```python
a = np.array([[1, 1, 1, 1], [1, 4, 16, 4**3], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([y1, y4, y10, y15])
x = scipy.linalg.solve(a, b)
x = np.around(x, 3)
print(a)
print(b)
print("веса - ",x)
a1 = np.array([1, 4, 10, 15])
b1 = np.array([y1, y4, y10, y15])
plt.plot(a1, b1)

#c0 + c1 * x^1 + c2 * x^2 + c3 * x^3 = f(1)
#c0 + c1 * x^1 + c2 * x^2 + c3 * x^3 = f(4)
#c0 + c1 * x^1 + c2 * x^2 + c3 * x^3 = f(10)
#c0 + c1 * x^1 + c2 * x^2 + c3 * x^3 = f(15)
```
