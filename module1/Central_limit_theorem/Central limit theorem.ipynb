# Центральная предельная теорема
## Рассмотрим гамма распределение.
$F(x) = x^{k-1} * \frac{e^{-x/\theta}}{\theta^k * Г(k)} ,    x > 0$


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
```

*Зададим гамма распределение k = 1, ${\theta}$ = 2. Сгенирируем выборку из 1000 случайный значений*

```python
gamma = sts.gamma(1, scale=2)
values = gamma.rvs(1000)
x = np.linspace(0, 10, 1000)
pdf = gamma.pdf(x)
```

*Построим гистограмму выборки и теоретическую плотность распределения нашей случайной величины*

```python
plt.plot(x, pdf)
plt.hist(values, bins=20, range=(0,10), density=True)
```

*Найдем мат. ожидание и дисперсию*

```python
k = 1 # Число степени свободы
t = 2 # Коэффициент масштаба
XE = k * t
XD = k * t**2
print("Мат. ожидание - " + str(XE))
print("Дисперсия - " + str(XD))
```

### *Оценим распределение выборочного среднего нашей случайной величины при выборке объема 5*

```python
n = 5
sample5 = np.array([])
for i in range(1000):
    sample_val5 = gamma.rvs(n)
    sample_mean = 0
    for elem in sample_val5:
        sample_mean += elem
    sample_mean /= n
    sample5 = np.append(sample5, sample_mean)

XE = 2
XD = 4  
sigma5 = math.sqrt(XD/n)
norm5 = sts.norm(XE, sigma5)
pdf5 = norm5.pdf(x[:500])
plt.plot(x[:500], pdf5, label='n = 5')
plt.hist(sample5, bins=50, range=(0,10), density=True)
plt.ylabel('доля выборки')
plt.xlabel('$x$')
```
