```python
import numpy as np
import pandas as pd
from sklearn.cluster import MeanShift
from collections import Counter
```

```python
df = pd.read_csv('datasets/umn_foursquare_datasets/checkins.dat', sep='|')
df = df.drop([0])
columns = []
for col in df.columns:
    columns.append(col.replace(' ', ''))
df.columns = columns
df.loc[:, 'latitude'] = pd.to_numeric(df.loc[:, 'latitude'], errors='coerce')
df.loc[:, 'longitude'] = pd.to_numeric(df.loc[:, 'longitude'], errors='coerce')
df = df.dropna(subset=['latitude'])
df.reset_index(drop=True)
```

```python
test = df.iloc[:100000, 3:5]
```

```python
clustering = MeanShift(bandwidth=0.1)
```

```python
clustering.fit(test)
```

```python
clust_arr = clustering.fit_predict(test)
```

```python
clust_dict = Counter(clust_arr)
for key in list(clust_dict.keys()):
    if clust_dict[key] <= 15:
        clust_dict.pop(key)
```

```python
coord = [[33.751277, -118.188740],[25.867736, -80.324116], [51.503016, -0.075479], [52.378894, 4.885084], [39.366487, 117.036146],[-33.868457, 151.205134]]
```

```python
ans = []
for point in list(clust_dict.keys()):
    x, y = clustering.cluster_centers_[point][0], clustering.cluster_centers_[point][1]
    for i in coord:
        dist = np.sqrt((x - i[0]) ** 2 + (y - i[1]) ** 2)
        ans.append((dist, x, y))
ans.sort()
```
